```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.plot(x,y, 'ro')
plt.show()
```

```python
plt.plot(x,y, 'ro-')
plt.show()
```

```python
def paint_xy_lines(x,y):
    for xi in x:
        plt.axvline(x=xi, color='gray', linestyle='--', linewidth=0.7)
    #print('x-> ',xi)
    for yi in y:
        plt.axhline(y=yi, color='gray', linestyle='--', linewidth=0.7) 
    #print('y-> ',yi)
```

```python
print(x)
print(y)
plt.plot(x,y, 'yD:')
for xi in x:
    plt.axvline(x=xi, color='gray', linestyle='--', linewidth=0.7)
    print('x-> ',xi)
for yi in y:
    plt.axhline(y=yi, color='gray', linestyle='--', linewidth=0.7) 
    print('y-> ',yi)   
plt.show()
```
