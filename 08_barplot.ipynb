```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
countries = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1300, 40, 200, 48, 24]
```

```python
plt.bar(countries,population,width=0.5, color=['red','blue'])
plt.xticks(np.arange(5), ('India','Japon','Mexico','Colombia', 'Alemania'), rotation=45)
plt.show()
```

```python
plt.barh(countries,population)
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)

ax1.bar(countries,population,width=0.5, color=['red','blue'],label='Population by country')
ax1.set_xticks(np.arange(5), ('India','Japon','Mexico','Colombia', 'Alemania'), rotation=45)
ax1.set_xlabel('Countries')
ax1.set_ylabel('Population')
ax1.legend()
ax2.barh(countries,population)
ax2.set_yticks(np.arange(5), ('India','Japon','Mexico','Colombia', 'Alemania'), rotation=45)
ax2.set_xlabel('Population')
ax2.set_ylabel('Countries')
plt.tight_layout()
```
