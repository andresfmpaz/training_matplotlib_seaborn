```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

plt.subplot(1, 2, 1) numero de files ,numero de columnas , el index se usara el 1 el primero

```python
print(x)
print(y)
plt.subplot(1, 2, 1)
plt.plot(x,y, 'ro-')
plt.subplot(1, 2, 2)
plt.hist(y)
plt.show()
```

```python
print(x)
print(y)
plt.subplot(1, 2, 1)
plt.plot(x,y, 'ro-')
plt.plot(y,x, 'b:')
plt.subplot(1, 2, 2)
plt.pie(y)
plt.show()
```

```python
print(x)
print(y)
plt.subplot(2, 1, 1)
plt.plot(x,y, 'ro-')
plt.plot(y,x, 'b:')
plt.subplot(2, 1, 2)
plt.pie(y)
plt.show()
```
