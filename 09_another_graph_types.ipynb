```python
data = np.random.randint(1,50, 100)
data
```

Histogram

```python
plt.hist(data, bins=50, width=1,histtype='bar')
plt.show()
```

bloxplot


```python
plt.boxplot(data,vert=False, patch_artist = True,notch = True)
plt.show()
```

showfliers = True is for exclude data that will introduce a bias, you can see 200 is outside of the boxplot and if you set up showfliers = False , boxplot will include it

```python
data_withflier = np.append(data,200) # 200 is so different of the other ranger of data
plt.boxplot(data_withflier,vert=False, patch_artist = True,notch = True, showfliers= True)
print(data_withflier.mean())
plt.show()
```

SCATTER dispersion

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30* np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area, c= colors,marker = 'o', alpha=0.5)
plt.show()
```

```python
x1 = np.random.rand(N)
y1 = np.sin(x1)
c1 = np.sin(x1)
```

```python

plt.scatter(x1,y1, s=area, c= c1,marker = 'o', alpha=0.7)
plt.show()
```
