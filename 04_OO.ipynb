```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
z = x ** 3
print('x: ',x)
print('y: ',y)
print('z: ',z)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes.plot(x,y,'b')
plt.show()

```

```python
left = 0.1
botton = 0.1
width = 0.5
height = 0.9
fig = plt.figure()
axes = fig.add_axes([left, botton, width, height])
axes_2 = fig.add_axes([left+width, botton, 0.4, 0.3])
axes_2.set_facecolor('gray')

axes.plot(x,y,'b')
axes_2.plot(y,x,'r')
plt.show()
```
