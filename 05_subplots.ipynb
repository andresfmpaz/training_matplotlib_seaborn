```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 20)
y = np.sin(x)
```

```python
print(x)
print('y:', y)
fig, axes = plt.subplots()
axes.plot(x,'b')
```

```python
print(x)
print('y:', y)
fig, axes = plt.subplots()
axes.plot(y,'r')
```

```python
print(x)
print('y:', y)
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
print(x)
print('y:', y)
fig, axes = plt.subplots()
axes.plot(y,x,'b')
plt.show()
```

```python
print(x)
print('y:', y)
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(y,x,'b')
axes[0].plot(x,'bo')
axes[1].plot(x,y,'r')
plt.show()
```

```python
print(x)
print('y:', y)
fig, (axes1,axes2) = plt.subplots(nrows=1,ncols=2)
axes1.plot(y,x,'b')
axes1.plot(x,'bo')
axes2.plot(x,y,'r')
plt.show()
```

```python
print(x)
print('y:', y)
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(nrows=2,ncols=4)
fig.suptitle('donde aparece')
ax1.set_title('ax1 title')
ax1.plot(np.sin(x),'b')
ax2.plot(np.cos(x),'bo:')
ax3.plot(np.tan(x),'r--')
ax4.plot((1/np.tan(x)),'r')

ax5.set_title('ax1 title')
ax5.plot(x,'b:')
ax6.plot(np.cos(x),'bo:')
ax7.plot(np.tan(x),'r--')
ax8.plot((1/np.tan(x)),'r')

fig.tight_layout()
for ax in fig.get_axes():
    ax.set(xlabel='x-label', ylabel='y-label')
plt.show()
```
